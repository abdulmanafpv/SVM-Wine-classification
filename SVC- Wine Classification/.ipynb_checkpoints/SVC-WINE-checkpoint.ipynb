```python
import pandas as pd
import numpy as np
import sklearn.model_selection as ms
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
import matplotlib.pyplot as plt
wine=pd.read_csv('C:\\Users\\user\\Desktop\\datas\\Wine1.data')
wine
```

```python
X=wine.drop(['Alcohol'],axis=1).values
Y=wine['Alcohol'].values
X1=wine[['Proline','Colour_Intensity']].values
X1
```

```python
X_train,X_test,Y_train,Y_test=ms.train_test_split(X1,Y,\
                    test_size=0.3,random_state=22)
X_train.shape
```

We can use SVM classifier

```python
SVMclassifier=svm.SVC()
Cval=2.**np.arange(-1,1.2,step=0.2)
print(Cval)
```

```python
n_grid=[{'C':Cval}]
cv_svc=GridSearchCV(estimator=SVMclassifier,
        param_grid=n_grid,cv=ms.KFold(n_splits=100))
cv_svc.fit(X_train,Y_train)
```

```python
best_c=cv_svc.best_params_['C']
print("The best parameter is :",best_c)
```

```python
svc_clf=svm.SVC(C=best_c)
svc_clf.fit(X_train,Y_train)
y_p=svc_clf.predict(X_test)
```

```python
print(metrics.classification_report(y_p,Y_test))
```

```python
C=best_c
svc=svm.SVC(kernel='linear',C=C).fit(X_train,Y_train)
rbf_svc=svm.SVC(kernel='rbf',gamma=0.7,C=C).fit(X_train,Y_train)
poly_svc=svm.SVC(kernel='poly',degree=3,C=C).fit(X_train,Y_train)
lin_svc=svm.LinearSVC(C=C).fit(X_train,Y_train)
```

```python
y_predict=svc.predict(X_test)
print(metrics.classification_report(y_predict,Y_test))
```

```python
y_p_rbf=rbf_svc.predict(X_test)
print(metrics.classification_report(y_p_rbf,Y_test))
```

```python
y_p_poly=poly_svc.predict(X_test)
print(metrics.classification_report(y_p_poly,Y_test))
```

```python
y_p_lin_svc=lin_svc.predict(X_test)
print(metrics.classification_report(y_p_lin_svc,Y_test))
```

We can Visualize the Data

```python
titles=['SVC','Linear SVC(linear kernel)','SVC with RBF Kernel','SVC with polynomial(degree 3)']
for i, clf in enumerate((svc,lin_svc,rbf_svc,poly_svc)):
    plt.subplot(2,2,i+1)
    plt.subplots_adjust(wspace=0.2,hspace=0.4)
    Z=clf.predict(np.c_[xx.ravel(),yy.ravel()])
    Z=Z.reshape(xx.shape)

    plt.contour(xx,yy,Z,cmap=plt.cm.coolwarm,alpha=0.8)
    plt.scatter(X1[:,0],X1[:,1],c=Y,cmap='jet')
    plt.xlabel("Proline")
    plt.ylabel("Colour_intensity")
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])
plt.show()
```
